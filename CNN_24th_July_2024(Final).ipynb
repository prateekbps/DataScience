# CNN model without using padding and strides

```python
model = Sequential()
model.add(Conv2D(16, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(16, kernel_size=(3,3), activation='relu', padding='valid'))
model.add(Flatten())
model.add(Dense(16, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.summary()
```

# Applying padding method

```python
model = Sequential()
# CNN 1
model.add(Conv2D(16, kernel_size=(3,3), activation='relu', padding='same', input_shape=(28,28,1)))
# CNN 2
model.add(Conv2D(16, kernel_size=(3,3), activation='relu', padding='same'))
# Flatten (2D to 1D)
model.add(Flatten())
# DNN
model.add(Dense(16, activation='relu'))
# output
model.add(Dense(10, activation='softmax'))
model.summary()
```

```python
import tensorflow as tf
from tensorflow import keras
# Data Augmentation package/library
from keras.preprocessing.image import ImageDataGenerator
from keras.utils import img_to_array, array_to_img, load_img
import matplotlib.pyplot as plt
```

```python
# importing the image

image_file = '/content/dog_images.jpeg'
dog_image = keras.preprocessing.image.load_img(image_file)
dog_image
```

```python
dog_array = keras.preprocessing.image.img_to_array(dog_image)
dog_array
```

```python
# Defining original image vs Transformed image

def plot_image(original_image, updated_image):
  plt.figure(figsize=(15,8))
  # original image
  plt.subplot(1,2,1)
  plt.title("Original Image")
  plt.imshow(original_image)

  # Transformed image
  plt.subplot(1,2,2)
  plt.title("Transformed Image")
  plt.imshow(updated_image)

  plt.show()
```

```python
# Flip an image
transform_fn = keras.preprocessing.image.ImageDataGenerator(horizontal_flip=False, vertical_flip=True, rotation_range=45)
transformed_array = transform_fn.random_transform(dog_array)
transformed_image = keras.preprocessing.image.array_to_img(transformed_array)
plot_image(dog_image, transformed_image)
```

# Pooling Layer - Adding MaxPooling Layer to reduce parameter

```python
from keras.layers import MaxPooling2D
```

```python
model = Sequential()
# CONV Layer and Relu layer
model.add(Conv2D(25, kernel_size=(3,3), strides=2, padding='same', activation='relu', input_shape=(28,28,1)))
# Max Pooling Layer
model.add(MaxPooling2D(pool_size=(2,2),strides=2, padding='valid'))
model.add(Conv2D(25, kernel_size=(3,3), strides=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2), strides=2, padding='valid'))
# Flatten means 2D to 1D
model.add(Flatten())
# Fully Connected Layer
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.summary()
```

```python
(x_train, y_train),(x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train.astype('float64')
x_test = x_test.astype('float64')
x_train /=255.0
x_test /=255.0
```

```python
x_train.shape
```

```python
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
```

image.png

```python
model = Sequential()
# conv layer 1
model.add(Conv2D(6, kernel_size=(5,5), padding='valid', strides=1,activation='relu', input_shape=(28,28,1)))
# Avg pooling - replace with MaxPooling
model.add(MaxPooling2D(pool_size=(2,2), strides=2, padding='valid' ))

# conv layer 2
model.add(Conv2D(16, kernel_size=(5,5), activation='relu'))
# pooling
model.add(MaxPooling2D(pool_size=(2,2), strides=2))

# Flatten
model.add(Flatten())

# Dense Layer - Fully Connected Layer (Model - MLP) - FC 1
model.add(Dense(120, activation='relu'))

# Fully Connected Layer 2
model.add(Dense(84, activation='relu'))
# output
model.add(Dense(10, activation='softmax'))

# summary
model.summary()
```

```python
# Compile the model
model.compile(optimizer='Adam', loss=keras.losses.categorical_crossentropy, metrics=['accuracy'])
```

```python
# Fit the model
history = model.fit(x_train, y_train, epochs=10, batch_size=64, validation_data=(x_test, y_test))
```
